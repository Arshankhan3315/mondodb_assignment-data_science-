Question-1- What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?

Answer-1-MongoDB is a popular and widely used open-source NoSQL (non-relational) database management system. It stores data in flexible, JSON-like documents with a dynamic schema, known as BSON (Binary JSON). MongoDB is designed to be scalable, high-performance, and easy to use for both small and large-scale applications.

Non-relational databases, often referred to as NoSQL databases, are databases that do not follow the traditional relational database model based on tables, rows, and columns. Instead, they use various data models like documents, key-value pairs, wide-column stores, or graphs to organize and represent data.

# Scenarios where MongoDB is preferred over SQL databases:

Semi-structured or unstructured data: MongoDB is well-suited for handling unstructured or semi-structured data because of its flexible schema. This is particularly useful when the data structure is evolving or unknown in advance.

Scalability and Performance: MongoDB is designed for horizontal scaling and can handle large amounts of data and high-velocity read and write operations. It can distribute data across multiple servers and handle a high volume of traffic efficiently.

Big Data and Real-time Analytics: In applications that require real-time analytics and handling of big data, MongoDB can be a good choice due to its ability to quickly process and analyze large volumes of data.

Rapid Development and Prototyping: MongoDB's schema-less design allows for faster development and prototyping, as changes in the data model can be accommodated without requiring a predefined schema.

Agile and Iterative Development: MongoDB fits well with agile development methodologies where frequent updates and changes to the data model are expected during the development process.

Single View Applications: Applications that require a single view of the data, especially where data is nested or hierarchical, benefit from MongoDB's document-based data storage.

Internet of Things (IoT): MongoDB is used extensively in IoT applications where it can handle a large volume of data generated by various devices and sensors.



Question-2. State and Explain the features of MongoDB.

Answer-2-MongoDB, a popular NoSQL database, offers several features that contribute to its widespread use and versatility

# Flexible Data Model (BSON):
MongoDB stores data in a flexible, schema-less format called BSON (Binary JSON). BSON allows for a rich and varied data model, including nested documents, arrays, and other data types, providing great flexibility in data representation.

# High Performance:
MongoDB is designed for high performance, with features like in-memory computing, support for indexing, and efficient query execution. It can handle a large volume of read and write operations, making it suitable for high-velocity applications.

# Scalability:
MongoDB is horizontally scalable, enabling seamless distribution of data across multiple servers or nodes in a cluster. This scalability allows for handling large amounts of data and increased performance as the application's load grows.

# Ease of Use and Development:
MongoDB is known for its ease of use and developer-friendly features. Its query language is intuitive and resembles JSON, making it easy for developers to work with. The lack of a fixed schema simplifies the development process and accelerates prototyping.

# Document-Oriented Storage:
MongoDB stores data in a document-oriented manner, where related information is stored together in a document. This approach allows for efficient retrieval of data and better performance, as it minimizes the need for complex joins often seen in relational databases.

# Indexing and Searching:
MongoDB supports indexing to improve query performance. Indexes can be created on any field in a document, and MongoDB's query optimizer uses these indexes to efficiently locate and retrieve the requested data.
